```python
df = pd.get_dummies(df, columns=['genre','key','mode','time_signature'])
```

```python
import numpy as np

df['split'] = np.random.randn(df.shape[0], 1)

msk = np.random.rand(len(df)) <= 0.67

train = df[msk]
test = df[~msk]

train = train.sample(frac=1)
test = test.sample(frac=1)
```

```python
test = test.drop(columns=['artist_name','track_name','split'])
```

```python
test.info()
```

```python
train = train.drop(columns=['artist_name','track_name','split'])
```

```python
from sklearn import preprocessing
import matplotlib
import seaborn as sns
from sklearn.compose import ColumnTransformer
%matplotlib inline
matplotlib.style.use('ggplot')

scaler = preprocessing.RobustScaler()
scaler2 = preprocessing.RobustScaler()

#num_train = train[['popularity','acousticness','danceability','duration_ms','energy','instrumentalness','liveness','loudness','speechiness','tempo','valence']]
#num_test = test[['popularity','acousticness','danceability','duration_ms','energy','instrumentalness','liveness','loudness','speechiness','tempo','valence']]

ct = ColumnTransformer([
        ('somename', scaler, list(test.columns))],
         remainder='passthrough')

s_train = ct.fit_transform(train)
s_train = pd.DataFrame(s_train, index = train.index, columns = train.columns)
    
s_test = ct.fit_transform(test)
s_test = pd.DataFrame(s_test, index = test.index, columns = test.columns)
    
#s_train = scaler.fit_transform(num_train)
#s_train = pd.DataFrame(s_train, index = train.index, columns = num_train.columns)

#s_test = scaler.fit_transform(num_test)
#s_test = pd.DataFrame(s_test, index = test.index, columns = num_test.columns)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import LinearRegression
from sklearn import svm




X = s_train.drop(columns='popularity')
Y = s_train['popularity']

# fit the model
clf = SGDRegressor(loss = "epsilon_insensitive",max_iter=200000)
clf.fit(X, Y)



```

```python
linearreg = LinearRegression() #applying the linear regression model
linearreg.fit(X, Y) #fitting the model to the training data
```

```python
from sklearn.svm import SVR
svr = SVR(C=1, epsilon=0.2, gamma = 'auto', cache_size=2000)

ss_train = s_train.sample(n=50000)

XA = ss_train.drop(columns='popularity')
YA = ss_train['popularity']

svr.fit(XA, YA)
```

```python
from sklearn import metrics


X_test = s_test.drop(columns='popularity')
Y_test = s_test['popularity']



y_predict2 = clf.predict(X_test) #making predictions based on the testing unit
print(np.sqrt(metrics.mean_squared_error(Y_test, y_predict2))) #calculating the RMSE number

y_predict = linearreg.predict(X_test) #making predictions based on the testing unit
print(np.sqrt(metrics.mean_squared_error(Y_test, y_predict))) #calculating the RMSE number
```

```python


y_predict3 = svr.predict(X_test) #making predictions based on the testing unit
print(np.sqrt(metrics.mean_squared_error(Y_test, y_predict3))) #calculating the RMSE number
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor

sss_train = s_train

XAA = sss_train.drop(columns='popularity')
YAA = sss_train['popularity']

regr =  ExtraTreesRegressor(n_estimators = 1000, max_depth=2, random_state=0, n_jobs =-1,max_features = None)
regr.fit(XAA, YAA)

y_predict4 = regr.predict(X_test) #making predictions based on the testing unit
print(np.sqrt(metrics.mean_squared_error(Y_test, y_predict4))) #calculating the RMSE number

importances = regr.feature_importances_
std = np.std([tree.feature_importances_ for tree in regr.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]])
plt.show()
```
